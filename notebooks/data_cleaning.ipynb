rainfall dataset stats 

```python
# By city
```

```python
rainfall_stats_cities = rainfall_df.groupby("City").agg(
    Average_Rainfall=("Rainfall - (MM)", "mean"),
    Maximum_Rainfall=("Rainfall - (MM)", "max"),
    Minimum_Rainfall=("Rainfall - (MM)", "min"),
)

print(rainfall_stats_cities)
```

```python
rainfall_stats_provinces = rainfall_df.groupby("Province").agg(
    Average_Rainfall=("Rainfall - (MM)", "mean"))

print(rainfall_stats_provinces)
```

```python
rainfall_stats_months = rainfall_df.groupby("Month").agg(
    Average_Rainfall=("Rainfall - (MM)", "mean"))

print(rainfall_stats_months)
```

```python
rainfall_stats_seasons = rainfall_df.groupby("Quarter").agg(
    Average_Rainfall=("Rainfall - (MM)", "mean"))

print(rainfall_stats_seasons)
```

Weather dataset stats

```python
weather_stats_cities = weather_df.groupby("city").agg(
    Average_Max_Temp=("temp_max", "mean"),
    Average_Min_Temp=("temp_min", "mean"),
    Hottest_Day=("temp_max", "max"),
    Coldest_Day=("temp_min", "min")
)

print(weather_stats_cities)
```

```python
weather_stats_month = weather_df.groupby("month").agg(
    Average_Max_Temp=("temp_max", "mean"),
    Average_Min_Temp=("temp_min", "mean"),
    Hottest_Day=("temp_max", "max"),
    Coldest_Day=("temp_min", "min")
)

print(weather_stats_month)
```

```python
avg_solar = weather_df["solar_radiation"].mean()

print("Average Solar Radiation:", round(avg_solar, 2))
```

```python
monthly_solar = weather_df.groupby("month")["solar_radiation"].mean().round(2)

print(monthly_solar)
```

```python
highest_month = monthly_solar.idxmax()
lowest_month = monthly_solar.idxmin()

print("Highest:", highest_month, "-", monthly_solar.max())
print("Lowest:", lowest_month, "-", monthly_solar.min())
```
